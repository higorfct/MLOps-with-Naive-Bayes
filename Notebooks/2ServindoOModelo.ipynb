#Convertendo as variáveis string para variáveis numéricas

```python
# Percorre todas as colunas do DataFrame 'credito'
for col in credito.columns:
    # Verifica se o tipo da coluna é 'object' (geralmente texto/categórico)
    if credito[col].dtype == 'object':
        # Converte a coluna para o tipo 'category' e depois transforma as categorias em códigos numéricos
        credito[col] = credito[col].astype('category').cat.codes
```
